```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import *
```

```python

t = 2
s = 0.001
n = 20/s

x = []
f = []
g = []

f.append(0)
x.append(0)
g.append(1)

fn = f[0] + 0.5*g[0]*s
gn = g[0] + (2 - 2*f[0])*s
f.append(fn)
g.append(gn)
x.append(s)

while t <= n:
    fn = f[t-1] + 0.5*g[t-1]*s
    f.append(fn)
    gn= g[t-1] + (2-2*f[t-1])*s
    g.append(gn)
    xn = x[t-1] + s
    x.append(xn)
    t +=1
    
#print(f)
#print(g)
#print(x)
```

```python
plt.plot(x, f, color='green')
plt.plot(x, g, color='blue')
plt.xlabel('x')
plt.ylabel('f')
plt.axis('tight')
plt.show()
```

```python
t = 0
print('Корни уравнения f(x) = g(x), (x,y):')
while t<=n:
    if math.fabs(f[t] - g[t]) <= 0.001:
        print(round(x[t],2), round(f[t],2))
    t += 1
```
